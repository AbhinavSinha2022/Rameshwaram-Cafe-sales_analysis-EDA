## Average Item price by Item_type

```python
sales = df.groupby(['item_name'],as_index=False)['item_price'].sum().sort_values(by='item_price',ascending=False)
sns.barplot(x='item_name',y='item_price',data= sales,palette='hls',hue='item_name')
```

```python
sales = df.groupby(['item_name'],as_index=False)['item_price'].sum().sort_values(by='item_price',ascending=False)
sns.boxplot(x='item_name',y='item_price',data= sales,palette='hls',hue='item_name')
```

### Transaction Amount and Transaction Type

```python
sales = df.groupby(['t_type'],as_index=False)['t_amount'].sum().sort_values(by='t_amount',ascending=False)
sns.barplot(x='t_type',y='t_amount',data= sales,palette='crest',hue='t_amount')
```

```python
sales = df.groupby(['t_type'],as_index=False)['t_amount'].sum().sort_values(by='t_amount',ascending=False)
sns.violinplot(x='t_type',y='t_amount',data= sales,palette='crest',hue='t_amount')
```

#### Most Transaction Happend in Cash

## Sales By Date or Time

```python
sales = df.groupby(['date'],as_index=False)['sale_time'].sum().sort_values(by='sale_time',ascending=False).head(10)
sns.barplot(x='date',y='sale_time',data= sales,palette='hls',hue='sale_time')
```

```python
sales = df.groupby(['date'],as_index=False)['t_amount'].sum().sort_values(by='t_amount',ascending=False).head(10)
sns.barplot(x='date',y='t_amount',data= sales,palette='plasma',hue='t_amount')
```

#### From the above graph we can see that most revenue was generated on 3/19/2023

```python
sales = df.groupby(['date'],as_index=False)['t_amount'].sum().sort_values(by='t_amount',ascending=False).tail(10)
sns.barplot(x='date',y='t_amount',data= sales,palette='plasma',hue='t_amount')
```

#### From the above graph we can see that least revenue was generated on 11/09/2022 and and 09/23/2022

```python
sales = df.groupby(['date'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=False).head(10)
sns.barplot(x='date',y='quantity',data= sales,palette='dark',hue='quantity')
```

#### From the above graph we can see that most sale happend on 10-11-2022 and quantity sold was 71

```python
sales = df.groupby(['date'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=True).tail(10)
sns.barplot(x='date',y='quantity',data= sales,palette='dark',hue='quantity')
```

#### From the above graph we can see that most sale happend on 02-03-2022 and quantity sold was 52

# Sales by Salesperson

```python
sales = df.groupby(['received_by'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=False)
sns.barplot(x='received_by',y='quantity',data= sales,palette='inferno',hue='quantity')
```

```python
sales = df.groupby(['received_by'],as_index=False)['item_price'].sum().sort_values(by='item_price',ascending=False)
sns.barplot(x='received_by',y='item_price',data= sales,palette='dark',hue='item_price')
```

#### From the Above two graphs we can see that male employess have sold more both in number of item_price and Quantity

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='received_by', y='item_price', data=df, estimator=sum, errorbar=None,hue='received_by', palette='dark')
plt.title('Total Sales by Salesperson')
plt.xlabel('Salesperson')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
